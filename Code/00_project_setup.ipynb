```python
from pathlib import Path
import sys
import subprocess
import importlib.util

ROOT = Path.cwd()
if ROOT.name == "Code":
    ROOT = ROOT.parent

DATA = ROOT / "DATA"
SCRIPTS = ROOT / "scripts"

print("Python executable:", sys.executable)
print("Repo root:", ROOT)
print("DATA exists:", DATA.exists())
print("scripts exists:", SCRIPTS.exists())
```

```python
required_packages = ["torch", "torchvision", "sklearn", "pandas", "numpy", "matplotlib", "PIL"]

for pkg in required_packages:
    found = importlib.util.find_spec(pkg) is not None
    print(f"{pkg:12s} -> {'OK' if found else 'MISSING'}")
```

```python
raw_expected = [
    DATA / "raw" / "isic2019" / "ISIC_2019_Training_Input.zip",
    DATA / "raw" / "isic2019" / "ISIC_2019_Test_Input.zip",
    DATA / "raw" / "milk10k" / "MILK10k_Training_Input.zip",
    DATA / "raw" / "milk10k" / "MILK10k_Test_Input.zip",
]

processed_expected = [
    DATA / "processed" / "isic2019" / "train.csv",
    DATA / "processed" / "isic2019" / "val.csv",
    DATA / "processed" / "isic2019" / "test.csv",
    DATA / "processed" / "milk10k" / "external_val.csv",
    DATA / "processed" / "class_weights.json",
    DATA / "processed" / "metadata_ohe.pkl",
    DATA / "processed" / "metadata_scaler.pkl",
]
```

```python
print("RAW DATA CHECK")
missing_raw = [p for p in raw_expected if not p.exists()]

if missing_raw:
    print("Missing raw files:")
    for p in missing_raw:
        print(" -", p.relative_to(ROOT))
else:
    print("All expected raw files exist.")
```

```python
if missing_raw:
    print("\nRunning download_data.py ...")
    subprocess.run([sys.executable, str(SCRIPTS / "download_data.py")], check=True)
else:
    print("\nSkipping download_data.py (raw files already exist).")
```

```python
print("PROCESSED DATA CHECK")
missing_processed = [p for p in processed_expected if not p.exists()]

if missing_processed:
    print("Missing processed files:")
    for p in missing_processed:
        print(" -", p.relative_to(ROOT))
else:
    print("All expected processed files exist.")
```

```python
if missing_processed:
    print("\nRunning data_prep.py ...")
    subprocess.run([sys.executable, str(SCRIPTS / "data_prep.py")], check=True)
else:
    print("\nSkipping data_prep.py (processed files already exist).")
```

```python
print("\nFINAL CHECK")
for p in processed_expected:
    print(f"{p.relative_to(ROOT)} -> {'OK' if p.exists() else 'MISSING'}")
```
